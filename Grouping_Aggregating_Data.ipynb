## 2. Grouping Data with pandas

### 2.1 Grouping Subset Data 

```python
# group the subset data

df.groupby('product_name')
```

```python
 # group the “department_id" and apply mean values for the “order_number” 
    
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# aggregation without use of the agg() function

df.groupby('department_id')['order_number'].mean()
```

```python
# the use of column name within square brackets 

df.groupby('department_id').order_number.mean()
```

```python
# performing multiple aggregations

df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

### 2.2 Apply Subset Groups Into The Main Data Frame *

```python
# performing multiple aggregations as on subset data

ords_prods_merge.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

### Analyze the result. How do the results for the entire dataframe differ from those of the subset. 

#### The mean values are slightly different in many department categories but generally the sample means are quite close to actual means as on the entire data frame. The means are usually greater in the data frame which covers the entire data. This is expected as there are more than 31M  more data rows in the main data frame and therefore data values are expected to be more widely spread out. However, there are few execptions, the mean for department 1, 2, 6, 16, 18 and 21 are greater in the subset data frame. The difference between the department id 2 and 21 can be explained due to the fact that the max order numbers are 96 and 98 in subset whereas they are 99 on the entire data frame. As the subset data frame includes "only" 1M rows, and it is a sample it won't be equal to the entire dataframe, and therefore the sample mean might not be systematically smaller or larger than the entire data frame. 

```python
# performing multiple aggregations - means, max and min for prices based on the product name

ords_prods_merge.groupby('product_name').agg({'prices': ['mean', 'min', 'max']})
```

```python
# performing multiple aggregations - mean for days based on the product name (min and max are known)

ords_prods_merge.groupby('product_name').agg({'orders_day_of_week': ['mean']})
```

### 2.3 Creating the Loyalty flag -  Aggregating Data for Different Customer Types

```python
# establish flag criteria - create “max_order” column, group “user_id” column, generate maximum orders on “order_number”

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

### 2.5 Spending Categories (above and below average price per product) *

#### Create a spending flag for each user based on the average price across all their orders using the following criteria: If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.” If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# establish spending flag criteria - create mean column, group “user_id” column, generate mean for “prices”

ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# check the data 

ords_prods_merge.head()
```

```python
# deriving columns with loc()

ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'Spender_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'Spender_flag'] = 'Low spender'
```

```python
ords_prods_merge['Spender_flag'].value_counts(dropna = False)
```

### 2.6 Frequent and Non-Frequent Customers * 

#### To determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows: If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.” If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer." If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
ords_prods_merge.columns
```

```python
# establish order frequency flag criteria - create median column for order frequency, group “user_id” column, generate median for “prices”

ords_prods_merge['median_order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# deriving columns with loc()

ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] > 20, 'Frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order_frequency'] <= 20) & (ords_prods_merge['median_order_frequency'] > 10), 'Frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] <= 10, 'Frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['Frequency_flag'].value_counts(dropna = False)
```
